```python
for col in df.columns[1:]:
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_xticks(df['week'])

    # put a green line on proficient
    ax.axhline(3, color='green', linestyle='--')

    ax.set_xticklabels(df['week'])
    ax.set_xlabel('Week')
    ax.set_ylabel('Level')
    ax.set_ylim(-0.5, 4.5)
    ax.set_yticks([0, 1, 2, 3, 4])
    ax.set_yticklabels(['Undefined', 'Orienting', 'Beginning', 'Proficient', 'Advanced'])

    # put colors on the yticklabels
    colors = {
        0: 'grey',
        1: 'red',
        2: 'orange',
        3: 'green',
        4: 'blue'
    }
    for i, label in enumerate(ax.get_yticklabels()):
        label.set_color(colors[i])

    plt.tight_layout()
    plt.title(f"'{col}' progress from week 1 to week 12")

    # fill area below the line depending on the y level
    for i, row in df.iterrows():
        if i == 11:
            break
        current_level = row[col]
        next_level = df.at[i+1, col]

        ax.fill_between(
            x=[row['week'], row['week']+1],
            y1=-0.5,
            y2=row[col],
            alpha=0.4,
            color=colors[row[col]]
        )

    # plot the line as a step
    ax.step(df['week'], df[col], where='post', color='black', marker='o')

    plt.savefig(f'progress_{col}.png', dpi=300, bbox_inches='tight')
    plt.show()
```
